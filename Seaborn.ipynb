```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sns.set(style='ticks', color_codes=True)
```

```python
# Load a sample data set
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.describe()
```

```python
# Plot a category plot - using stripplot with jitter set to True
sns.stripplot(x='day', y='total_bill', jitter=True, data=tips)
```

```python
# Swarm plot
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
## Set male and female for the color
sns.swarmplot(x='day', y='total_bill', hue = 'sex', data=tips)
```

```python
# Swarmplot by smoker or non smoker
sns.swarmplot(x="smoker", y="tip", order=["No", "Yes"], data=tips);
```

```python
# Create a box plot by day
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
# Box plot with smoker vs non-smoker
sns.boxplot(x='day', y='total_bill', hue = 'smoker', data=tips)
# Move the legend outside the figure
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
# Boxenplot - similar to box plot but shows distribution
sns.boxenplot(x='day', y='total_bill', hue = 'smoker', data=tips)
# Move the legend outside the figure
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.__version__
```

```python
# Violin plot

sns.violinplot(x='day', y='total_bill',  hue = 'smoker', data=tips)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
# Violin plot

sns.violinplot(x='day', y='total_bill', color='teal', hue = 'smoker', data=tips)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
# Count plot to show the count for each group
sns.countplot(x='day', hue = 'smoker', data=tips)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
import numpy as np
import pandas as pd
```

```python
sales = pd.read_excel("C:\\Users\\adity\\Documents\\TableauWork\\Customer_SuperStore.xlsx")
```

```python
sales
```

```python
## Build a model using Linear Regression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
df_X = pd.DataFrame(sales['Sales'].values)
df_Y = pd.DataFrame(sales['Profit'].values)
model.fit(df_X, df_Y)
```

```python
## See the LR Coefficient generated by the model
model.coef_
```

```python
## See the LR Intercept generated by the model
model.intercept_
```

```python
## Predict profit using the sales values in the data source
sales_predict = model.predict(df_X)
sales_predict
```

```python
## Add a new column so we can build the chart
sales.insert(3,'Predicted profit', sales_predict)
```

```python
## Rearrange the columns
sales = sales[['Customer Name', 'Sales', 'Profit', 'Predicted profit']]
sales
```

```python
## Plot the data with the fitted line
import matplotlib.pyplot as plt
plt.scatter(sales['Sales'], sales['Profit'],  color='orange')
plt.plot(sales['Sales'], sales['Predicted profit'], color='red', linewidth=3)
plt.show()
```

```python
## Create a function that wraps the prediction
def predictProfit(sales):
    return model.predict(sales).tolist()
```

```python
## Test the function
predictProfit(2000)
```

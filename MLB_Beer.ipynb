```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
beer_file = "C:\\Users\\adity\\Documents\\DataScience\\mlb_beer_prices.csv"
```

```python
beers = pd.read_csv(beer_file)
```

```python
beers.describe()
```

```python
beers.head()
```

```python
sns.boxplot(x='year', y = 'price_per_ounce', data=beers)
```

```python
sns.boxenplot(x='year', y = 'price_per_ounce', data=beers)
```

```python
## See the trend over the years
sns.relplot(x='year', y = 'price_per_ounce', kind = 'line', hue = 'nickname', data=beers)
```

```python
# Price per ounce by size
sns.boxplot(x='size', y = 'price_per_ounce', data=beers)
```

```python
# Price by size
sns.boxplot(x='size', y = 'price', data=beers)
```

```python
# Price by city
plt.figure(figsize = (20,20))
sns.boxplot(x='city', y = 'price_per_ounce', data=beers)
plt.xticks(rotation=45)
```
